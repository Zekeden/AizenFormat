```python
import json
import os
import shutil

# Paths
coco_train_images = '/content/not_dog'
coco_val_images = '/content/val2017'
annotations_path = '/content/annotations'

train_output_dog = '/content/dataset/train/dog'
train_output_not_dog = '/content/dataset/train/not_dog'
val_output_dog = '/content/dataset/val/dog'
val_output_not_dog = '/content/dataset/val/not_dog'

# Create output directories
for folder in [train_output_dog, train_output_not_dog, val_output_dog, val_output_not_dog]:
    os.makedirs(folder, exist_ok=True)

def extract_images(annotation_file, images_dir, dog_dir, not_dog_dir):
    with open(annotation_file) as f:
        data = json.load(f)

    # Category id for dog in COCO is 18
    dog_category_id = 18

    # Get image ids with dogs
    dog_image_ids = set()
    for ann in data['annotations']:
        if ann['category_id'] == dog_category_id:
            dog_image_ids.add(ann['image_id'])

    # Map image_id to file_name
    image_id_to_filename = {img['id']: img['file_name'] for img in data['images']}

    # Copy images to dog or not_dog folders
    for image_id, filename in image_id_to_filename.items():
        src_path = os.path.join(images_dir, filename)
        if image_id in dog_image_ids:
            dst_path = os.path.join(dog_dir, filename)
        else:
            dst_path = os.path.join(not_dog_dir, filename)

        shutil.copy(src_path, dst_path)
        print(f"Copied {filename} to {'dog' if image_id in dog_image_ids else 'not_dog'} folder")

# Extract training images
extract_images(
    annotation_file=os.path.join(annotations_path, 'instances_train2017.json'),
    images_dir=coco_train_images,
    dog_dir=train_output_dog,
    not_dog_dir=train_output_not_dog
)

# Extract validation images
extract_images(
    annotation_file=os.path.join(annotations_path, 'instances_val2017.json'),
    images_dir=coco_val_images,
    dog_dir=val_output_dog,
    not_dog_dir=val_output_not_dog
)
```

```python
import os
import shutil
import json
from PIL import Image

def extract_and_crop_dogs(annotation_file, images_dir, dog_dir, not_dog_dir, threshold=0.3):
    with open(annotation_file) as f:
        data = json.load(f)

    dog_category_id = 18
    image_id_to_filename = {img['id']: img['file_name'] for img in data['images']}

    dog_boxes = {}
    for ann in data['annotations']:
        if ann['category_id'] == dog_category_id:
            dog_boxes.setdefault(ann['image_id'], []).append(ann['bbox'])

    for image_id, filename in image_id_to_filename.items():
        src_path = os.path.join(images_dir, filename)
        print(f"Processing {filename}...")

        if not os.path.exists(src_path):
            print(f"File {src_path} does not exist, skipping")
            continue

        image = Image.open(src_path).convert('RGB')
        img_area = image.width * image.height

        if image_id in dog_boxes:
            total_dog_area = sum([box[2]*box[3] for box in dog_boxes[image_id]])
            dog_ratio = total_dog_area / img_area
            print(f"Dog area ratio for {filename}: {dog_ratio:.3f}")

            if dog_ratio >= threshold:
                print(f"Keeping {filename} in dog_dir, cropping dogs")
                for idx, box in enumerate(dog_boxes[image_id]):
                    x, y, w, h = map(int, box)
                    dog_crop = image.crop((x, y, x + w, y + h))
                    crop_filename = f"{os.path.splitext(filename)[0]}_{idx}.jpg"
                    dog_crop.save(os.path.join(dog_dir, crop_filename))
                    print(f"Saved crop {crop_filename}")
                # Optionally, remove original large image after cropping:
                # os.remove(src_path)
            else:
                print(f"Moving {filename} to not_dog_dir (dog too small)")
                shutil.move(src_path, os.path.join(not_dog_dir, filename))
        else:
            print(f"Moving {filename} to not_dog_dir (no dog annotation)")
            shutil.move(src_path, os.path.join(not_dog_dir, filename))


# Example call:
extract_and_crop_dogs(
    annotation_file='/content/annotations/instances_train2017.json',
    images_dir='/content/dataset/train/dog',
    dog_dir='/content/dataset/train/dog',
    not_dog_dir='/content/dataset/train/not_dog',
    threshold=0.4
)
```

```python
import os
import json
from PIL import Image

val_dog_dir = '/content/dataset/val/dog'
annotations_file = '/content/annotations/instances_val2017.json'  # Update if your annotations file path differs

# Load annotations
with open(annotations_file) as f:
    data = json.load(f)

dog_category_id = 18

# Map image_id to file_name
image_id_to_filename = {img['file_name']: img['id'] for img in data['images']}

# Map image_id to dog bounding boxes
dog_boxes = {}
for ann in data['annotations']:
    if ann['category_id'] == dog_category_id:
        dog_boxes.setdefault(ann['image_id'], []).append(ann['bbox'])

for filename in os.listdir(val_dog_dir):
    if filename not in image_id_to_filename:
        print(f"Skipping {filename} - not in annotations")
        continue

    image_id = image_id_to_filename[filename]
    image_path = os.path.join(val_dog_dir, filename)
    image = Image.open(image_path).convert('RGB')

    if image_id in dog_boxes:
        boxes = dog_boxes[image_id]
        for idx, box in enumerate(boxes):
            x, y, w, h = map(int, box)
            cropped = image.crop((x, y, x + w, y + h))
            crop_filename = f"{os.path.splitext(filename)[0]}_crop{idx}.jpg"
            cropped.save(os.path.join(val_dog_dir, crop_filename))
        # After cropping all dog boxes, delete original image
        os.remove(image_path)
        print(f"Cropped and removed original: {filename}")
    else:
        print(f"No dog boxes found in {filename}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models, datasets, transforms
from torch.utils.data import DataLoader, random_split
from tqdm import tqdm  # for progress bar

# Use GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')

# Data transforms
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),      # Augmentation
    transforms.RandomRotation(10),          # Augmentation
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],  # ImageNet stats
        std=[0.229, 0.224, 0.225]
    )
])

val_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# Load dataset with train transforms initially (we'll split)
full_dataset = datasets.ImageFolder(root='/content/dataset/train', transform=train_transform)

# Split dataset into train and validation (80/20)
train_size = int(0.8 * len(full_dataset))
val_size = len(full_dataset) - train_size
train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])

# Override val dataset transform to validation transforms
val_dataset.dataset.transform = val_transform

# Create DataLoaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)

# Load pretrained ResNet50 with new API
from torchvision.models import resnet50, ResNet50_Weights

weights = ResNet50_Weights.DEFAULT
model = resnet50(weights=weights)

# Modify the last fully connected layer for binary classification (dog vs not dog)
num_features = model.fc.in_features
model.fc = nn.Linear(num_features, 2)
model = model.to(device)

# Loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

# Training loop with validation
num_epochs = 5
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    loop = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Train]")
    for imgs, labels in loop:
        imgs, labels = imgs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(imgs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * imgs.size(0)
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += imgs.size(0)

        loop.set_postfix(loss=running_loss / total, accuracy=correct / total)

    train_loss = running_loss / total
    train_acc = correct / total

    # Validation phase
    model.eval()
    val_loss = 0.0
    val_correct = 0
    val_total = 0

    with torch.no_grad():
        val_loop = tqdm(val_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Val]")
        for imgs, labels in val_loop:
            imgs, labels = imgs.to(device), labels.to(device)
            outputs = model(imgs)
            loss = criterion(outputs, labels)

            val_loss += loss.item() * imgs.size(0)
            _, preds = torch.max(outputs, 1)
            val_correct += (preds == labels).sum().item()
            val_total += imgs.size(0)

            val_loop.set_postfix(val_loss=val_loss / val_total, val_acc=val_correct / val_total)

    val_loss /= val_total
    val_acc = val_correct / val_total

    print(f"Epoch {epoch+1} Summary: Train Loss {train_loss:.4f}, Train Acc {train_acc:.4f}, Val Loss {val_loss:.4f}, Val Acc {val_acc:.4f}")

print("Training completed.")

# Save the trained model
torch.save(model.state_dict(), 'dog_detector.pth')
print("Model saved as dog_detector.pth")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import models, transforms
from PIL import Image

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Rebuild model
model = models.resnet50(weights=None)
num_features = model.fc.in_features
model.fc = nn.Linear(num_features, 2)

# Load weights
model.load_state_dict(torch.load('dog_detector.pth', map_location=device))
model = model.to(device)
model.eval()

print("Model loaded successfully")
```

```python
transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485,0.456,0.406],
        std=[0.229,0.224,0.225]
    )
])

class_names = ['dog', 'not_dog']
DOG_INDEX = class_names.index('dog')
```

```python
def test_image(image_path, threshold=0.7):
    img = Image.open(image_path).convert('RGB')
    img = transform(img).unsqueeze(0).to(device)

    with torch.no_grad():
        outputs = model(img)
        probs = F.softmax(outputs, dim=1)

    dog_prob = probs[0][DOG_INDEX].item()
    predicted_class = class_names[torch.argmax(probs).item()]

    print(f"Dog probability: {dog_prob:.4f}")
    print(f"Raw prediction: {predicted_class}")

    if dog_prob >= threshold:
        print("✅ FINAL: DOG")
        return "DOG"
    else:
        print("❌ FINAL: NOT DOG")
        return "NOT_DOG"
```
